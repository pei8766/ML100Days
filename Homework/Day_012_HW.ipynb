```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
a = pd.read_csv('boston.csv', usecols=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"])
print(a)
a.boxplot()
#Ans:TAX
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
b = pd.read_csv('boston.csv', usecols=["NOX", "DIS"])
print(b)
b.plot.scatter(x='NOX',y='DIS')
#Ans:負相關
```
